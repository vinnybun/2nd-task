```python
# Re-run this cell
import pandas as pd

# Read in the data
schools = pd.read_csv(file_link)

# Preview the data
schools.head()
```

```python
# Start coding here...
# NYC schools with the best math results
#subset the school_name and the average_math results
school_math = schools[["school_name", "average_math"]]
school_math

#filter the best math result (80% of 800)
math_filter = school_math[school_math["average_math"] >= (0.8 * 800)]
math_filter

#sort the best math results in decreasing order
best_math_schools = math_filter.sort_values("average_math" , ascending = False)
best_math_schools
```

```python
#top 10 performing schools
#create a  new column (total_SAT)
schools["total_SAT"] = schools[["average_math", "average_reading", "average_writing"]].sum(axis = 1)

#subset the school name and the total_SAT of the schools sorted to get the top 10
top_schools = schools[["school_name", "total_SAT"]].sort_values("total_SAT", ascending = False)

#top 10 schools
top_10_schools = top_schools.head(10)
top_10_schools
```
